```python
course_information = []
index = 0

while index < len(course_names):
    try:
        url = course_names.iloc[index].tolist()[0]
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        name = soup.find_all('dd')[0].get_text()
        street = soup.find_all('dd')[1].get_text()
        city = soup.find_all('dd')[2].get_text()
        zipcode = soup.find_all('dd')[3].get_text()
        county = soup.find_all('dd')[4].get_text()
        email = soup.find_all('dd')[5].get_text()
        website = soup.find_all('a', href=True)[12]['href']
        phone = soup.find_all('dd')[7].get_text()

        #append the data frame
        course_information.append((name, street, city, zipcode, county, email, website, phone))
        index +=1
        if index%100 == 0:
            print(index)
    except IndexError:
        index +=1
```

```python
df = pd.DataFrame(course_information)
```
